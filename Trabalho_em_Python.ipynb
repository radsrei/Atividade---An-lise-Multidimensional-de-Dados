```python
import sqlite3

# Conectar ao banco de dados (ou criar um novo se não existir)
conn = sqlite3.connect('meu_banco_de_dados.db')
cursor = conn.cursor()

# DDL para criar as tabelas
cursor.execute('''
CREATE TABLE Estudante (
    estudanteID INTEGER PRIMARY KEY,
    nome VARCHAR(100) NOT NULL,
    curso VARCHAR(100) NOT NULL
);
''')

cursor.execute('''
CREATE TABLE Instrutor (
    instrutorID INTEGER PRIMARY KEY,
    curso VARCHAR(100) NOT NULL
);
''')

cursor.execute('''
CREATE TABLE Aula (
    aulaID INTEGER PRIMARY KEY,
    Instituicao VARCHAR(100) NOT NULL,
    cidade VARCHAR(100) NOT NULL,
    estado CHAR(2) NOT NULL
);
''')

cursor.execute('''
CREATE TABLE Aulas_assistidas (
    estudanteID INTEGER,
    instrutorID INTEGER,
    aulaID INTEGER,
    notas DECIMAL(5,2),
    FOREIGN KEY (estudanteID) REFERENCES Estudante(estudanteID),
    FOREIGN KEY (instrutorID) REFERENCES Instrutor(instrutorID),
    FOREIGN KEY (aulaID) REFERENCES Aula(aulaID)
);
''')

# Commitar as mudanças e fechar a conexão
conn.commit()
conn.close()

print("Tabelas criadas com sucesso no 'meu_banco_de_dados.db'!")
```

### Consulta A: Estudantes com notas acima de 65 em aulas em SC, onde o curso do instrutor é diferente do curso do estudante.

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('meu_banco_de_dados.db')
cursor = conn.cursor()

query_a = """
SELECT e.nome, a.Instituicao, f.notas
FROM Aulas_assistidas f
JOIN Estudante e ON f.estudanteID = e.estudanteID
JOIN Instrutor i ON f.instrutorID = i.instrutorID
JOIN Aula a ON f.aulaID = a.aulaID
WHERE a.estado = 'SC'
  AND i.curso <> e.curso
  AND f.notas > 65;
"""

cursor.execute(query_a)
results_a = cursor.fetchall()

df_a = pd.DataFrame(results_a, columns=['Nome Estudante', 'Instituição Aula', 'Notas'])
display(df_a)

conn.close()
```

### Consulta B: Média de notas por estudante e instrutor em Joinville.

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('meu_banco_de_dados.db')
cursor = conn.cursor()

query_b = """
SELECT e.nome AS Aluno, i.instrutorID, AVG(f.notas) AS media_notas
FROM Aulas_assistidas f
JOIN Estudante e ON f.estudanteID = e.estudanteID
JOIN Aula a ON f.aulaID = a.aulaID
JOIN Instrutor i ON f.instrutorID = i.instrutorID
WHERE a.cidade = 'Joinville'
GROUP BY e.nome, i.instrutorID;
"""

cursor.execute(query_b)
results_b = cursor.fetchall()

df_b = pd.DataFrame(results_b, columns=['Aluno', 'ID Instrutor', 'Média Notas'])
display(df_b)

conn.close()
```

### Consulta C: Média de notas por instrutor em Joinville.

### Consulta F: Média de notas por estado, cidade e instituição com ROLLUP.

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('meu_banco_de_dados.db')
cursor = conn.cursor()

# SQLite não suporta diretamente WITH ROLLUP como a sintaxe mostrada.
# Uma alternativa é usar UNION ALL para simular o ROLLUP ou agrupar em diferentes níveis.
# Para fins de demonstração, vou simular o ROLLUP em SQLite com UNION ALL.

# A consulta original com WITH ROLLUP que causou o erro.
# query_f_sqlite = """
# SELECT
#     a.estado,
#     a.cidade,
#     a.Instituicao,
#     AVG(f.notas) AS media_notas
# FROM Aulas_assistidas f
# JOIN Aula a ON f.aulaID = a.aulaID
# GROUP BY a.estado, a.cidade, a.Instituicao WITH ROLLUP;
# """

# Vamos usar a sintaxe alternativa para simular o ROLLUP com UNION ALL no SQLite.
query_f_fixed = """
SELECT
    a.estado,
    a.cidade,
    a.Instituicao,
    AVG(f.notas) AS media_notas
FROM Aulas_assistidas f
JOIN Aula a ON f.aulaID = a.aulaID
GROUP BY a.estado, a.cidade, a.Instituicao

UNION ALL

SELECT
    a.estado,
    a.cidade,
    NULL AS Instituicao, -- Subtotal por cidade e estado
    AVG(f.notas) AS media_notas
FROM Aulas_assistidas f
JOIN Aula a ON f.aulaID = a.aulaID
GROUP BY a.estado, a.cidade

UNION ALL

SELECT
    a.estado,
    NULL AS cidade, -- Subtotal por estado
    NULL AS Instituicao,
    AVG(f.notas) AS media_notas
FROM Aulas_assistidas f
JOIN Aula a ON f.aulaID = a.aulaID
GROUP BY a.estado

UNION ALL

SELECT
    NULL AS estado, -- Total geral
    NULL AS cidade,
    NULL AS Instituicao,
    AVG(f.notas) AS media_notas
FROM Aulas_assistidas f
JOIN Aula a ON f.aulaID = a.aulaID
ORDER BY estado DESC, cidade DESC, Instituicao DESC; -- Ordena para ter os totais no final ou onde fizer mais sentido
"""

cursor.execute(query_f_fixed)
results_f = cursor.fetchall()

df_f = pd.DataFrame(results_f, columns=['Estado', 'Cidade', 'Instituição', 'Média Notas'])
display(df_f)

conn.close()
```

### Consulta G: Média de notas por curso de estudante e curso de instrutor com ROLLUP, combinado com UNION.

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('meu_banco_de_dados.db')
cursor = conn.cursor()

# SQLite não suporta diretamente WITH ROLLUP. Precisamos simular com UNION ALL.
query_g_fixed = """
-- Agregação por curso de estudante e curso de instrutor
SELECT
    e.curso AS curso_estudante,
    i.curso AS curso_instrutor,
    AVG(f.notas) AS media
FROM Aulas_assistidas f
JOIN Estudante e ON f.estudanteID = e.estudanteID
JOIN Instrutor i ON f.instrutorID = i.instrutorID
GROUP BY e.curso, i.curso

UNION ALL

-- Subtotal por curso de estudante (instrutor é NULL)
SELECT
    e.curso AS curso_estudante,
    NULL AS curso_instrutor,
    AVG(f.notas) AS media
FROM Aulas_assistidas f
JOIN Estudante e ON f.estudanteID = e.estudanteID
JOIN Instrutor i ON f.instrutorID = i.instrutorID
GROUP BY e.curso

UNION ALL

-- Subtotal por curso de instrutor (estudante é NULL)
SELECT
    NULL AS curso_estudante,
    i.curso AS curso_instrutor,
    AVG(f.notas) AS media
FROM Aulas_assistidas f
JOIN Estudante e ON f.estudanteID = e.estudanteID
JOIN Instrutor i ON f.instrutorID = i.instrutorID
GROUP BY i.curso

UNION ALL

-- Total geral (ambos NULL)
SELECT
    NULL AS curso_estudante,
    NULL AS curso_instrutor,
    AVG(f.notas) AS media
FROM Aulas_assistidas f
JOIN Estudante e ON f.estudanteID = e.estudanteID
JOIN Instrutor i ON f.instrutorID = i.instrutorID
ORDER BY curso_estudante ASC, curso_instrutor ASC;
"""

cursor.execute(query_g_fixed)
results_g = cursor.fetchall()

df_g = pd.DataFrame(results_g, columns=['Curso Estudante', 'Curso Instrutor', 'Média'])
display(df_g)

conn.close()
```
